```python
import matplotlib.pyplot as plt

# Datos
valores = [88, 792]
nombres = ['Viñas Sustentables', 'Viñas Tradicionales']

# Crear gráfico circular
plt.figure(figsize=(8, 8))
plt.pie(valores, labels=nombres, autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Gráfico Circular de Dos Variables')

# Mostrar el gráfico
plt.show()
```
