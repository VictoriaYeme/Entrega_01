```python
import matplotlib.pyplot as plt

# Datos
categorias = ['Viñedo', 'Suelo', 'Nutricional', 'Malezas', 'Enfermedades', "Productos", "Residuos", "Vigor", "Riego", "Energia", "Capacitación", "Salud", "Biodiversidad", "Sustentabilidad", "Reclamaciones"]
puntaje = [16, 26, 48, 48, 48, 24, 36, 32, 68, 72, 32, 92, 20, 12, 28]

# Crear el gráfico de barras
plt.bar(categorias, puntaje, width=0.5)

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Puntaje')
plt.title('Evaluación')

# Ajustar el espaciado entre las etiquetas del eje x
plt.xticks(rotation=90)

# Mostrar el gráfico
plt.show()
```

```python
# HTML
!pip install plotly

import matplotlib.pyplot as plt
import plotly.express as px

# Datos
categorias = ['Viñedo', 'Suelo', 'Nutricional', 'Malezas', 'Enfermedades', "Productos", "Residuos", "Vigor", "Riego", "Energia", "Capacitación", "Salud", "Biodiversidad", "Sustentabilidad", "Reclamaciones"]
puntaje = [16, 26, 48, 48, 48, 24, 36, 32, 68, 72, 32, 92, 20, 12, 28]

# Crear el gráfico de barras
plt.bar(categorias, puntaje, width=0.5)

# Etiquetas y título
plt.xlabel('Categorías')
plt.ylabel('Puntaje')
plt.title('Evaluación')

# Ajustar el espaciado entre las etiquetas del eje x
plt.xticks(rotation=90)

# Exportar el gráfico a un archivo HTML con Plotly
fig = px.bar(x=categorias, y=puntaje, labels={'x': 'Categorías', 'y': 'Puntaje'}, title='Evaluación')
fig.write_html('grafico_Gallet.html')
```
