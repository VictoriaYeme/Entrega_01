```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Crear un DataFrame con los datos proporcionados
data = {'Categoría': ['Superficie Vitícola fuera de la RM', 'Bodegas fuera de la Región Metropolitana',
                      'Trabajo fuera de la Región Metropolitana', 'Viñas en O\'Higgins', 'Viñas en El Maule'],
        'Porcentaje': [91.0, 92.0, 85.0, 32.7, 37.9]}
df = pd.DataFrame(data)

# Ordenar el DataFrame por Porcentaje en orden descendente
df = df.sort_values(by='Porcentaje', ascending=False)

# Configurar el estilo de la visualización
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# Crear el gráfico de barras horizontal
barplot = sns.barplot(x='Porcentaje', y='Categoría', data=df, palette='viridis')

# Agregar etiquetas y título
plt.xlabel('Porcentaje')
plt.ylabel('Categoría')
plt.title('Porcentajes en Diferentes Categorías')

# Mostrar el gráfico
plt.show()
```

```python
import plotly.express as px
import pandas as pd

# Datos proporcionados
data = {
    'Categoría': ['Bodegas Activas', 'Productores', 'Empresas Exportadoras', 'Pymes en Empresas Exportadoras'],
    'Cantidad': [800, 11697, 394, 299]
}

df = pd.DataFrame(data)

# Asignar colores
colors = ['blue', 'orange', 'green', 'red']

# Crear gráfico de dispersión
fig = px.scatter(
    df,
    x='Categoría',
    y='Cantidad',
    color='Categoría',
    size='Cantidad',
    size_max=50,
    color_discrete_sequence=colors,
    title='Cantidad por Categoría',
    labels={'Categoría': 'Categoría', 'Cantidad': 'Cantidad'},
    log_y=True  # Configurar el eje y en escala logarítmica
)

# Mostrar el gráfico
fig.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
bodegas_activas = 800
porcentaje_bodegas_fuera_rm = 92.0
total_bodegas = 17000

# Configurar el tamaño de la figura
plt.figure(figsize=(10, 6))

# Plot Bodegas Activas
plt.bar(0, bodegas_activas, color='blue', label='Bodegas Activas')

# Plot Porcentaje de Bodegas fuera de la Región Metropolitana
plt.bar(1, porcentaje_bodegas_fuera_rm, color='orange', label='Bodegas fuera de la RM (%)')

# Plot Total de Bodegas
plt.bar(2, total_bodegas, color='green', label='Total de Bodegas')

# Configurar ejes y leyenda
plt.xticks([0, 1, 2], ['Bodegas Activas', 'Bodegas fuera de la RM (%)', 'Total de Bodegas'])
plt.ylabel('Cantidad')
plt.title('Datos de Bodegas')
plt.legend()

# Mostrar el gráfico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos proporcionados
labels = ['Bodegas Activas', 'Bodegas fuera de la RM', 'Bodegas en la RM', 'Bodegas']
values = [800, 92.0, 8, 17000]
percentages = [100, 92.0, 8, 17]

# Crear un arreglo de colores utilizando la paleta 'viridis'
colors = plt.get_cmap('viridis')(np.linspace(0, 1, len(labels)))

# Configurar el tamaño del gráfico
plt.figure(figsize=(10, 6))

# Dibujar un gráfico de barras apiladas
plt.bar(labels, values, color=colors, alpha=0.7)

# Agregar etiquetas y porcentajes
for i, (value, percentage) in enumerate(zip(values, percentages)):
    plt.text(i, value + 100, f"{percentage}%", ha='center', va='bottom', fontsize=10)

# Configurar el título y las etiquetas
plt.title('Distribución de Bodegas', fontsize=14)
plt.xlabel('Categoría', fontsize=12)
plt.ylabel('Cantidad', fontsize=12)

# Mostrar el gráfico
plt.show()
```
